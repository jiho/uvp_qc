```python
import pandas as pd

# read the data
d = pd.read_csv("cast_xb_datfile.txt", sep=";", header=None, )

# get the series of abundances
s = d[14] 

# define outliers
k = 8      # size of one side of the moving window
tau = 0.75 # quantile to consider as correct values
mult = 0.6 # multiplier of the quantile under which points are considered outliers
d['outlier'] = s < mult * s.rolling(2*k+1, center=True, min_periods=k).quantile(tau)
```

```python
d['outlier_num'] = d['outlier'].astype(int)
d.plot.scatter(x=0, y=14, c='outlier_num', colormap='viridis', ylim=(0,3000), s=1)
```
